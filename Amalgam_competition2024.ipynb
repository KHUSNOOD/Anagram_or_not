```python

from sklearn.feature_selection import SelectKBest, chi2, f_regression

# Select top 5 features with high chi-squared values
selector = SelectKBest(f_regression, k=20)
X = selector.fit_transform(X,y)
```

```python
from sklearn.decomposition import PCA

# Initialize PCA with the number of components you want to retain
n_components = 8  # Number of principal components
pca = PCA(n_components=n_components)

# Fit PCA to the data
X = pca.fit_transform(X)

# Print the transformed data shape
print("Transformed data shape:", X.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_valid,y_train,y_valid= train_test_split(X,y , train_size=0.7, random_state=42)
```

```python
# model training
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

from sklearn.linear_model import LinearRegression
lr= LinearRegression()
model= lr.fit(X_train,y_train)
y_pred= model.predict(X_valid)
rmse= np.sqrt(mean_squared_error(y_pred,y_valid))
```

```python
rmse
```

```python
# model 2

from sklearn.linear_model import Ridge
ridge = Ridge(alpha = 1, solver = 'saga', tol = 1e-4, random_state = 42)
ridge.fit(X_train, y_train)
y_pred= ridge.predict(X_valid)
rmse= np.sqrt(mean_squared_error(y_pred,y_valid))
```

```python
rmse
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
rfr= RandomForestRegressor(random_state=42)
param_grid= {
    'n_estimators': [1,5,10,40,100,200,300,350],
    'max_features': [2,5,10,20,32,36,40,44]
}
grid_search= GridSearchCV(rfr, param_grid, cv=5,scoring='neg_mean_squared_error',
return_train_score=True)

# Fit the GridSearchCV object to the data
grid_search.fit(X_train, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)

# Get the best model
best_lr = grid_search.best_estimator_

# Make predictions on the test set using the best model
y_pred = best_lr.predict(X_valid)

# Evaluate the model using mean squared error
rmse = np.sqrt(mean_squared_error(y_valid, y_pred))


```

```python
# Import libraries
from sklearn.svm import SVR


# Define and train the SVM model
clf = SVR(kernel='linear', C=1.0)  # Linear kernel, adjust C for regularization
clf.fit(X_train, y_train)

# Make predictions on test data
y_pred = clf.predict(X_valid)

# Evaluate the model using mean squared error
rmse = np.sqrt(mean_squared_error(y_valid, y_pred))
```

```python
from sklearn.tree import DecisionTreeRegressor

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Initialize Decision Tree Regressor
regressor = DecisionTreeRegressor()

# Fit the model to the training data
regressor.fit(X_train, y_train)

# Predict on the testing set
y_pred = regressor.predict(X_valid)

# Calculate Mean Squared Error (MSE) as a performance metric
rmse = np.sqrt(mean_squared_error(y_valid, y_pred))
print("Root Mean Squared Error:", rmse)
```
